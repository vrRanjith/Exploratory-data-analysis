##### Let's see the total number of campaigns count by Company and Industry Type.

```python
campaign_grouped = data_frame.groupby(['Company', 'Industry'])['No of Campaign'].agg('sum')
campaign_grouped
```

##### Let's see city wise campaign count

```python
campaign_grouped_city = data_frame.groupby(['City', 'Company',])['No of Campaign'].agg('sum')
campaign_grouped_city
```

```python
# Let's see the top 10 cities visually which had the most number of campaigns counducted.

city_visual = data_frame.groupby('City')['No of Campaign'].agg('sum')
city_df = pd.DataFrame({'City':city_visual.index, 'campaign_count':city_visual.values})
city_df.sort_values(by='campaign_count', ascending=False, inplace=True)

x = city_df['City'][:10]
y = city_df['campaign_count'][:10]

plt.rcParams['figure.figsize'] = (18, 10)
plt.bar(x,y)
plt.show()
```

##### The above visual shows the top 10 cities with highest campaigns conducted, with mumbai ranking first, where all the top 10 cities are in india.

```python
# Let's do the same for the company.

Company_visual = data_frame.groupby('Company')['No of Campaign'].agg('sum')
Company_df = pd.DataFrame({'Company':Company_visual.index, 'campaign_count':Company_visual.values})
Company_df.sort_values(by='campaign_count', ascending=False, inplace=True)

x = Company_df['Company'][:10]
y = Company_df['campaign_count'][:10]

plt.rcParams['figure.figsize'] = (25, 6)
plt.bar(x,y)
plt.show()
```

##### The above visual shows the top 10 Companies with most campaigns conducted, with `Toothsi` ranking first, followed by `Bombay Shaving Company`, `Paharpur Cooling Towers`, `Corning`, `Jagatjit Industries`, `Lionsgate`, `Unnati`, `PUMA`, `Ralph Lauren`, `Birayani By Kilo`.

```python
# Let's do the same for the Industry.

Industry_visual = data_frame.groupby('Industry')['No of Campaign'].agg('sum')
Industry_df = pd.DataFrame({'Industry':Industry_visual.index, 'campaign_count':Industry_visual.values})
Industry_df.sort_values(by='campaign_count', ascending=False, inplace=True)

x = Industry_df['Industry'][:10]
y = Industry_df['campaign_count'][:10]

plt.rcParams['figure.figsize'] = (19, 6)
plt.bar(x,y)
plt.show()
```

##### From above visual we can see that `Dental Tech` industry had the highest number of campaigns count.

```python
# Finally let's visualise the top 10 team that had the most campaigns.
Team_visual = data_frame.groupby('Team')['No of Campaign'].agg('sum')
Team_df = pd.DataFrame({'Team':Team_visual.index, 'campaign_count':Team_visual.values})
Team_df.sort_values(by='campaign_count', ascending=False, inplace=True)

x = Team_df['Team'][:10]
y = Team_df['campaign_count'][:10]

plt.rcParams['figure.figsize'] = (19, 6)
plt.bar(x,y)
plt.show()
```

##### We can see that the `HR team` has the highest number of campaigns conducted overall.
